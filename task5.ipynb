```python
c=b.drop(['TMC','End_Lat', 'End_Lng','Description', 'Number','Airport_Code','Zipcode','Timezone','Weather_Timestamp',
       'Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)',
       'Visibility(mi)', 'Wind_Direction', 'Wind_Speed(mph)',
       'Precipitation(in)','Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight',
       'Astronomical_Twilight'],axis=1)
c.shape
```

```python
d= c.dropna(subset=['City','Weather_Condition'])
d.shape
```

```python
print(d.isnull().sum())
```

```python
print(d.head())
```

```python
# Plot the distribution of accidents by hour of the day


import matplotlib.pyplot as plt
d['Hour'] = d['Start_Time'].dt.hour

plt.figure(figsize=(10, 5))
d['Hour'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title('Accidents by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Number of Accidents')
plt.show()
```

1.The chart shows that the highest number of accidents occur around 7 AM and 4 PM to 5 PM.This suggests that rush hours, when people are commuting to and from work, are the times with the highest accident rates.                              
2.There is a significant peak around 7 AM, indicating that a large number of accidents happen during the morning rush hour.                                                 
3.Another peak is observed around 4 PM to 5 PM, which corresponds to the evening rush hour when people are returning home from work.                                
4.The early morning hours (midnight to 5 AM) and late evening hours (8 PM to midnight) have the lowest number of accidents.This can be attributed to lower traffic volumes during these times.

```python
# Plot the distribution of accidents by weather condition

plt.figure(figsize=(10, 5))
d['Weather_Condition'].value_counts().nlargest(10).plot(kind='bar', color='orange')
plt.title('Top 10 Weather Conditions During Accidents')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

1.The highest number of accidents occur during clear weather conditions, with over 800,000 accidents reported.             
2.The second-highest number of accidents occur during fair weather, with around 600,000 accidents.                       
3.Mostly cloudy conditions also see a significant number of accidents, with over 500,000 reported.                        
4.Overcast conditions account for a considerable number of accidents, with approximately 400,000 reported.               
5.These conditions have a moderate number of accidents, ranging from 300,000 to 400,000.                              
6.These conditions have fewer accidents, with scattered clouds and light rain accounting for around 200,000 to 300,000 accidents.
7.The fewest accidents occur during light snow and rain, with each condition accounting for fewer than 100,000 accidents.


```python
#SCATTER PLOT OF ACCIDENTS SPOTS IN COUNTRY

import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Start_Lng', y='Start_Lat', data=d, alpha=0.3)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

1.The Eastern United States, particularly the Northeast and areas along the East Coast, show a high density of accidents.

2.The western part of the country has a mix of urban and rural areas, with high traffic density in cities and along major highways.                                                    
3.The Southern United States, including Texas and Florida, also show dense clusters of accidents.Major cities like Houston, Dallas, and Miami contribute to the higher density of accidents.

```python
#PIE CHART OF DISRIBUTION OF ACCIDENTS DUE TO WEATHER CONDITIONS


import matplotlib.pyplot as plt

weather_counts = d['Weather_Condition'].value_counts().nlargest(10)
plt.figure(figsize=(10, 7))
weather_counts.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=140)
plt.title('Distribution of accients due to weather conditions')
plt.ylabel('')
plt.show()
```

1. clear weather conitions are the most common, making up 24.8 % of Distribution of accients due to  clear weather conditions.
2. fair weather conitions  come next, accounting for 16.8 % of the accidents due to  clear weather conditions.
3. light snow rain and light rain Types of accidents make up the remaining 1.3 % and 1.5% respectively.

```python
# BAR CHART OF TOP 20 CITIES OF NUMBER OF ACCIDENTS

accidents_by_street = d['City'].value_counts()
accidents_by_street[:20]
fig, ax = plt.subplots(figsize=(8,5))
accidents_by_street[:20].plot(kind='bar',color='green')
ax.set(title = 'Top 20 cities By Number of Accidents',
       xlabel = 'cities',
       ylabel = 'Accidents Count')
plt.show()
```

1.The highest number of accidents occur in  houston city with over 980,000 accidents reported.                                                              
2.The second-highest number of accidents  in los angeles city with around 790,000 accidents.                                                                                                                                                     
3.The fewest accidents occur  in portland city of top most 20 cities, with accounting for around 21000 accidents.
